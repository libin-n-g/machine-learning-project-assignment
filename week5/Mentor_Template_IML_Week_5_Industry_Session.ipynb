## **<font color='blue'>Tree Based Models**

Tree-based models are a class of machine learning algorithms that use a decision tree as their fundamental building block. These models are commonly used for both classification and regression tasks.

A decision tree is a flowchart-like structure where
- Each internal node represents a decision based on a feature.
- Each branch represents an outcome of the decision.
- And each leaf node represents a class label (for classification) or a predicted value (for regression).

image.png

###**Advantages**


Tree-based models have several advantages:

**Interpretability:** Decision paths in a tree are easily interpretable, making these models suitable for explaining predictions to stakeholders.

**Feature Importance:** Tree-based models can provide insights into feature importance, helping identify the most relevant variables.

**Handling Non-linearity:** They can capture complex, non-linear relationships between features and target variables.

**Robustness:** They handle outliers and missing values relatively well.

###**Disadvantages**

However, they also have limitations:

**Overfitting:** Trees can become overly complex and overfit the training data, leading to poor generalization.

**Bias to Dominant Features:** In some cases, dominant features can heavily influence the tree's decisions.

**Instability:** Small changes in data can lead to different trees, affecting stability.

Tree-based models are versatile tools for predictive modeling, allowing you to make accurate predictions and gain insights into the relationships within your data.

## **<font color='blue'>Classification and Regression Trees**

**Classification and Regression Trees (CART)** are powerful machine learning algorithms used for both classification and regression tasks. They create decision trees that partition the data into subsets based on feature conditions to predict categorical (class labels) or continuous (numeric) target variables.

Here's how Classification and Regression Trees work:

1. **Decision Tree Structure:**
A decision tree is a hierarchical structure consisting of nodes. The top node is called the root node, which poses a question about a feature. Each branch corresponds to a possible answer to the question, leading to other nodes or leaf nodes. Leaf nodes contain the final predictions or class labels.

2. **Splitting Criteria:**
For classification trees, the goal is to minimize impurity or increase homogeneity in each subset of data. Common impurity measures include Gini Impurity and Entropy, which quantify the disorder or purity of class labels in a subset. For regression trees, the goal is to minimize the variance of the target variable within each subset.

3. **Recursive Splitting:**
The decision tree-building process is recursive. At each node, the algorithm selects the best feature and its corresponding threshold to split the data into two or more subsets. This process continues until a stopping criterion is met, such as a maximum tree depth or a minimum number of samples in each leaf.

4. **Classification Trees:**
In a classification tree, the leaf nodes represent class labels. The majority class in a leaf node determines the predicted class for new data points that follow the same decision path.

5. **Regression Trees:**
In a regression tree, the leaf nodes represent predicted values. The average (or another metric) of the target variable values in a leaf node is used as the prediction for new data points that follow the same decision path.

### **Pruning**

Pruning is a vital technique in machine learning used to prevent overfitting and enhance model generalization. Overfitting occurs when a model becomes overly complex, fitting noise in the training data rather than capturing underlying patterns. Pruning involves selectively removing parts of a model that contribute little to predictive accuracy.

Pruning is commonly applied to decision trees to improve their performance and interpretability.

image.png

### **Cost-Complexity Pruning:**

Cost-Complexity Pruning involves introducing a complexity parameter (alpha) to control the trade-off between model accuracy and complexity. The formula for the cost-complexity criterion is:

![Cost-Complexity Criterion](https://latex.codecogs.com/svg.latex?R_\alpha(T)%20=%20R(T)%20+%20\alpha%20%7C%7C%20T%20%7C%7C)

Where:
- ![R(T)](https://latex.codecogs.com/svg.latex?R(T)) is the original loss function (e.g., Gini Impurity or Mean Squared Error) of the tree T.
- ![||T||](https://latex.codecogs.com/svg.latex?||T||) represents the complexity of the tree T (e.g., the number of nodes).

### **Pruning Process:**

1. **Tree Growth:** Initially, a decision tree is fully expanded to capture patterns in the data.

2. **Validation:** A validation dataset or cross-validation assesses the impact of pruning on model performance.

3. **Cost-Complexity Pruning:** Different values of alpha are considered, and the corresponding subtrees are evaluated based on the cost-complexity criterion.

4. **Pruning Decision:** The subtree with the smallest cost-complexity criterion is selected. This subtree provides the best trade-off between accuracy and complexity.

### **Benefits of Pruning:**

1. **Improved Generalization:** Pruning prevents overfitting, allowing the model to generalize better to new, unseen data.

2. **Interpretability:** Pruned models are simpler and more interpretable, making them easier to understand.

3. **Reduced Complexity:** Pruned models have fewer nodes, leading to faster predictions and lower memory usage.


## **<font color='blue'>Two families of Ensemble Methods**

1. **Averaging methods**, the driving principle is to build several estimators independently and then to average / vote  their predictions. On average, the combined estimator is usually better than any of the single base estimator because its variance is reduced.

  E.g. Bagging methods,Forests of randomized trees, ...

2. **Boosting methods**, base estimators are built sequentially and one tries to reduce the bias of the combined estimator. The motivation is to combine several weak models to produce a powerful ensemble.

  E.g. AdaBoost,Gradient Tree Boosting, ...

### **Ensemble Types:**

**Bagging (Bootstrap Aggregating):** This involves training multiple models on different bootstrap samples of the training data and averaging their predictions. Random Forest is an example of a bagging ensemble using decision trees.

**Boosting:** In boosting ensembles like AdaBoost or Gradient Boosting, each subsequent model focuses on correcting the errors of the previous ones, leading to a more accurate overall prediction.

**Stacking:** Stacking combines predictions from multiple models (not necessarily of the same type) using another model (meta-model) to make the final prediction.

## **<font color='blue'>Bagging (Bootstrap Aggregating)**

Bagging, which stands for Bootstrap Aggregating, is a powerful ensemble technique in machine learning that enhances model performance and robustness. It involves creating multiple subsets of the training data through random sampling with replacement and training individual models on these subsets. The final prediction is obtained by combining the predictions of these models.


**How Bagging Works:**

1. **Bootstrap Sampling:** Bagging involves creating multiple subsets (samples) of the training data by randomly sampling with replacement. Each subset is called a "bootstrap sample."

2. **Individual Models:** A base model, often a complex model like a decision tree, is trained independently on each bootstrap sample. This results in multiple individual models, each having learned from slightly different data.

3. **Aggregation of Predictions:** During prediction, the outputs of individual models are aggregated to make the final prediction. For classification tasks, this typically involves a majority vote among the class predictions of individual models. For regression tasks, predictions are averaged.

## **<font color='blue'>Random Forest:**

Random Forest is a versatile and powerful ensemble learning technique that combines multiple decision trees to create a more accurate and robust predictive model. It leverages the concept of bagging (Bootstrap Aggregating) and introduces additional randomness to each individual decision tree, resulting in improved generalization and reduced overfitting.

**How Random Forest Works:**

1. **Bootstrap Sampling:** Like bagging, Random Forest starts by creating multiple bootstrap samples (subsets) of the training data.

2. **Random Feature Selection:** In addition to using different data samples, Random Forest introduces randomness by selecting a random subset of features for each decision tree. This helps decorrelate the trees and makes the ensemble more diverse.

3. **Individual Decision Trees:** Multiple decision trees are trained independently on each bootstrap sample, with each tree using a different subset of features.

4. **Prediction Aggregation:** During prediction, the outputs of individual decision trees are aggregated to make the final prediction. For classification, this often involves majority voting; for regression, predictions are typically averaged.

## **<font color='blue'>Boosting**

Boosting is an ensemble machine learning technique that combines multiple weak learners (simple models) to create a single strong predictive model. Unlike other ensemble methods that build models independently and aggregate their predictions, boosting emphasizes improving the overall model iteratively by focusing on instances that were previously misclassified. This results in a final model with enhanced predictive accuracy.

**How Boosting Works**:

1. **Sequential Training:** Boosting trains a series of models sequentially, where each model builds on the knowledge of the previous ones.

2. **Weighted Training Data:** During each iteration, the training data is assigned weights. Instances that were misclassified in previous iterations are given higher weights, making them more influential in the subsequent model's training.

3. **Model Aggregation:** After all iterations, the predictions of individual models are combined to make the final prediction. The aggregation process often involves weighted majority voting for classification tasks or weighted averaging for regression tasks.

**Some of the boosting algorithms include:**

- **AdaBoost (Adaptive Boosting):** Adjusts instance weights at each iteration to focus on previously misclassified instances.
- **Gradient Boosting:** Uses gradient descent optimization to minimize errors and improve the model's performance.
- **XGBoost (Extreme Gradient Boosting):** A highly optimized gradient boosting algorithm known for its efficiency and performance.

**Key Concepts:**

- **Weak Learners:** Boosting uses weak learners, which are models that perform slightly better than random guessing. These can be simple decision trees with limited depth or other models with low complexity.

- **Error Reduction:** Each subsequent model focuses on correcting the errors made by the previous models, gradually improving the overall prediction accuracy.

## **<font color='blue'>AdaBoost**

AdaBoost, also known as Adaptive Boosting, is like a team of experts working together to solve a challenging problem. Each expert (model) specializes in certain aspects of the problem, but they all collaborate to provide the best possible solution.

**How Adaboost Works:**
1. **Learning from Mistakes:** AdaBoost trains a sequence of models one after another. Each model pays extra attention to the instances that the previous models struggled with. This way, the team learns from their collective mistakes.

2. **Weighted Voting:** During prediction, the experts (models) vote on the correct answer, but their votes are weighted based on their expertise. More accurate experts have a stronger say in the final decision.

3. **Challenging Instances:** AdaBoost identifies the instances that the team finds difficult and makes sure that subsequent models focus on improving predictions for those instances.

4. **Strong Ensemble:** By combining the insights of individual experts and learning from mistakes, AdaBoost creates a strong ensemble model that excels at solving the problem.

## **<font color='blue'>Gradient Boosting**

Gradient Boosting is a machine learning technique that assembles a strong predictive model by iteratively combining the outputs of multiple weak learners (typically decision trees). It focuses on minimizing errors by using gradient descent optimization to improve the model's performance step by step.

**How Gradient Boosting Works:**

1. **Sequential Training:** Gradient Boosting trains a sequence of weak learners, each building on the knowledge of its predecessors.

2. **Residual Learning:** Instead of using the original target values, each weak learner is trained to predict the residuals (differences between predicted and actual values) of the previous model.

3. **Weighted Data:** Misclassified instances are assigned higher weights, directing subsequent models to focus on correcting those errors.

4. **Gradient Descent:** The models are trained using gradient descent optimization, adjusting their parameters to minimize the errors.

5. **Model Aggregation:** The predictions of weak learners are combined, often through weighted averaging, to make the final prediction.

## **<font color='blue'>XGBoosting**

XGBoost is an advanced and highly optimized version of Gradient Boosting. It builds on the principles of Gradient Boosting and introduces additional features to enhance both performance and predictive accuracy.

**How XGBoosting Works:**

1. **Sequential Ensemble:** XGBoost trains a sequence of weak models, typically decision trees, in a boosting fashion. Each new model corrects the errors made by its predecessors.

2. **Residual Learning:** Similar to Gradient Boosting, XGBoost employs residual learning. It focuses on predicting the differences (residuals) between the actual and predicted values of the previous model.

3. **Gradient Optimization:** XGBoost optimizes its objective function by utilizing gradient descent. It calculates the gradient of the loss function and makes small, precise adjustments to model parameters to minimize errors.

4. **Regularization:** XGBoost introduces L1 (Lasso) and L2 (Ridge) regularization terms to prevent overfitting. These terms add penalties for large parameter values, encouraging simpler and more generalized models.

5. **Handling Missing Values:** XGBoost can gracefully handle missing data during training and prediction, reducing the need for extensive preprocessing.

6. **Parallel Processing:** To enhance efficiency, XGBoost supports parallel processing for model training, making use of all available resources.

7. **Cross-validation and Early Stopping:** XGBoost includes built-in cross-validation for parameter tuning and an early stopping mechanism to halt training when performance plateaus.

8. **Prediction Aggregation:** Finally, the predictions of all models are combined to produce the ensemble's final prediction.

##  Building the model

### Model evaluation criterion:

#### Model can make wrong predictions as:
1. Predicting a customer will buy the product and the customer doesn't buy - Loss of resources
2. Predicting a customer will not buy the product and the customer buys - Loss of opportunity

#### Which case is more important?

#### How to reduce this loss i.e need to reduce False Negatives?
* Company wants Recall to be maximized, greater the Recall lesser the chances of false negatives.

**Let's start by building different models using KFold and cross_val_score and tune the best model using GridSearchCV and RandomizedSearchCV**

- `Stratified K-Folds cross-validation` provides dataset indices to split data into train/validation sets. Split dataset into k consecutive folds (without shuffling by default) keeping the distribution of both classes in each fold the same as the target variable. Each fold is then used once as validation while the k - 1 remaining folds form the training set.

```python
models = []  # Empty list to store all the models

# Appending models into the list
models.append(("Bagging", BaggingClassifier(random_state=1)))
models.append(("Random forest", RandomForestClassifier(random_state=1)))
models.append(("GBM", GradientBoostingClassifier(random_state=1)))
models.append(("Adaboost", AdaBoostClassifier(random_state=1)))
models.append(("Xgboost", XGBClassifier(random_state=1, eval_metric="logloss")))

results = []  # Empty list to store all model's CV scores
names = []  # Empty list to store name of the models


# loop through all models to get the mean cross validated score
print("\n" "Cross-Validation Performance:" "\n")

for name, model in models:
    scoring = "recall"
    kfold = StratifiedKFold(
        n_splits=5, shuffle=True, random_state=1
    )  # Setting number of splits equal to 5
    cv_result = cross_val_score(
        estimator=model, X=X_train, y=y_train, scoring=scoring, cv=kfold
    )
    results.append(cv_result)
    names.append(name)
    print("{}: {}".format(name, cv_result.mean() * 100))

print("\n" "Training Performance:" "\n")

for name, model in models:
    model.fit(X_train, y_train)
    scores = recall_score(y_train, model.predict(X_train)) * 100
    print("{}: {}".format(name, scores))
```

```python
# Plotting boxplots for CV scores of all models defined above
fig = plt.figure(figsize=(10, 7))

fig.suptitle("Algorithm Comparison")
ax = fig.add_subplot(111)

plt.boxplot(results)
ax.set_xticklabels(names)

plt.show()
```

- We can see that XGBoost is giving the highest cross-validated recall followed by GBM
- The boxplot shows that the performance of Adaboost is consistent with no outlier
- We will tune - GBM, AdaBoost and XGBoost and see if the performance improves.

**First, let's create two functions to calculate different metrics and confusion matrix so that we don't have to use the same code repeatedly for each model.**

The following model performance classification function will return the accuracy, recall, precision and F1 score.

```python
# defining a function to compute different metrics to check performance of a classification model built using sklearn
def model_performance_classification_sklearn(model, predictors, target):
    """
    Function to compute different metrics to check classification model performance

    model: classifier
    predictors: independent variables
    target: dependent variable
    """

    # predicting using the independent variables
    pred = model.predict(predictors)

    acc = accuracy_score(target, pred)  # to compute Accuracy
    recall = recall_score(target, pred)  # to compute Recall
    precision = precision_score(target, pred)  # to compute Precision
    f1 = f1_score(target, pred)  # to compute F1-score

    # creating a dataframe of metrics
    df_perf = pd.DataFrame(
        {"Accuracy": acc, "Recall": recall, "Precision": precision, "F1": f1,},
        index=[0],
    )

    return df_perf
```

The following function will return the confusion matrix for a model

```python
def confusion_matrix_sklearn(model, predictors, target):
    """
    To plot the confusion_matrix with percentages

    model: classifier
    predictors: independent variables
    target: dependent variable
    """
    y_pred = model.predict(predictors)
    cm = confusion_matrix(target, y_pred)
    labels = np.asarray(
        [
            ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.flatten().sum())]
            for item in cm.flatten()
        ]
    ).reshape(2, 2)

    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=labels, fmt="")
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
```

###   GBM

Let's tune GBM model using **GridSearch**

We define certain hyperparameters in the grid and GridSearchCV build model using every possible combination of the hyperparameters defined in the grid and it returns the best combination of the hyperparameters. Grid Search is usually computationally expensive.

**Grid Search**

```python
%%time

# defining model
model = GradientBoostingClassifier(random_state=1)

# Parameter grid to pass in GridSearchCV

param_grid = {
    "n_estimators": np.arange(10, 110, 10),
    "learning_rate": [0.1, 0.01, 0.2, 0.05, 1],
    "max_depth": [1,2,3],
}

# Type of scoring used to compare parameter combinations
scorer = metrics.make_scorer(metrics.recall_score)

# Calling GridSearchCV
grid_cv = GridSearchCV(estimator=model, param_grid=param_grid, scoring=scorer, cv=5, n_jobs = -1)

# Fitting parameters in GridSearchCV
grid_cv.fit(X_train, y_train)

print(
    "Best Parameters:{} with CV score={}:".format(grid_cv.best_params_, grid_cv.best_score_)
)
```

```python
# building model with best parameters
gbm_tuned1 = GradientBoostingClassifier(
    n_estimators=60,
    learning_rate=1,
    max_depth=2,
)

# Fit the model on training data
gbm_tuned1.fit(X_train, y_train)
```

**Checking model performance**

```python
# Calculating different metrics on train set
GBM_grid_train = model_performance_classification_sklearn(
    gbm_tuned1, X_train, y_train
)

GBM_grid_test = model_performance_classification_sklearn(
    gbm_tuned1, X_test, y_test
)
print("Training performance:")
print(GBM_grid_train)

print("*************************************")

print("Test performance:")
print(GBM_grid_test)

print("*************************************")

# creating confusion matrix
confusion_matrix_sklearn(gbm_tuned1, X_train, y_train)
```

**Observations**
- On comparing the CV score and the training score, model is slightly overfitting.
